```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

### Data Preprocessing

```python
data = pd.read_excel('datafile.xls')
data.head()
```

```python
agri = data.iloc[0].to_numpy()[2:]
elec = data.iloc[8].to_numpy()[2:]
comm = data.iloc[21].to_numpy()[2:]
bank = data.iloc[19].to_numpy()[2:]
trans = data.iloc[13].to_numpy()[2:]
```

```python
years = data.columns[2:].to_list()
years = np.array([int(g.split('-')[-2]) for g in years])
```

### GDP from different sectors since 1950

```python
plt.scatter(years[:63], trans[:63], s=8, label='Transport')
plt.scatter(years[:63], agri[:63], s=8, label='Agriculture')
plt.scatter(years[:63], elec[:63], s=8, label='Electricity')
plt.scatter(years[:63], bank[:63], s=8, label='Bank')
plt.scatter(years[:63], comm[:63], s=8, label='Communication')
plt.title('GDP from different sectors')
plt.xlabel('Year')
plt.ylabel('GDP (in million crore rupees)')
plt.legend()
```

### GDP Growth in different sectors between 2003-12

```python
w = 0.1
fig = plt.figure(figsize=(15, 7))
plt.bar(years[-11:-1]-w, trans[-10:], width=w, label='Transport')
plt.bar(years[-11:-1]-2*w, agri[-11:-1], width=w, label='Agriculture')
plt.bar(years[-11:-1], elec[-11:-1], width=w, label='Electricity')
plt.bar(years[-11:-1]+w, bank[-10:], width=w, label='Bank')
plt.bar(years[-11:-1]+2*w, comm[-10:], width=w, label='Communication')
plt.title('% GDP Growth in different sectors between 2003-12')
plt.xlabel('Year')
plt.xticks(years[-11:-1])
plt.ylabel('% GDP Growth')
plt.legend()
```

### Sectorwise %GDP Growth since 1950

```python
plt.boxplot([comm[64:], agri[64:], elec[64:],  bank[64:-1], trans[64:-1]],\
            labels=['Communication','Agriculture', 'Electricity', 'Bank', 'Transport'])
plt.ylabel('% GDP Growth')
plt.xlabel('Sector')
plt.title('% GDP Growth in different sectors since 1950')
```
